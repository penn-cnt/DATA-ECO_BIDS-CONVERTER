# Example usages of update_datalake.py

This code is meant to help maintain a list of imaging keywords used in BIDS. Each site and each project may have its own protocol naming scheme, and this is a means to keep a running list of these options.

## View datalake

```python
%run update_datalake.py --datalake BIDS_DATALAKE.pickle --review
```

The code will prompt you for a substring to search for under protocol.

If you enter 'n', the code will exit.

If you enter 'y', it will enter into a loop to allow you to type in substrings to find a selection of the datalake.

 This alows you to find the protocol of interest. Lets look for anything that has 't2_' in its name.

```python
%run update_datalake.py --datalake BIDS_DATALAKE.pickle --review
```

## Update datalake

You can update the datalake in one of two ways. You can enter data manually, or you can provide a csv file with entries. (Sample csv files can be found in the examples folder.)

### Single Input

```python
%run update_datalake.py --datalake BIDS_DATALAKE.pickle --manual
```

### Bulk Input

```python
%run update_datalake.py --datalake BIDS_DATALAKE.pickle --upload --upload_file examples/example_upload_w_header.csv
```

# Deleting Entries

At present, the easiest method to delete or modify an entry is to either manually or upload a new value. The code will update the datalake with the more recent version. If you wish to delete something entirely, you can pass the protocol name a completely empty list of values and it will remove it from the datalake.

## Different Sites

These protocol names are defined for the HUP environment. Each site will likely have its own naming conventions. To this end, you can set the site using the command line argument --site , which will show you results for your site and help you start building your sites datalake.
